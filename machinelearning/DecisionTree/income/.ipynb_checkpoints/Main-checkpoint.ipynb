## import Libs and Frameworks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

## read dataset

```python
df = pd.read_csv('income.csv')
df.head()
```

## Clean data

```python
df = df.replace('?',np.nan).dropna().reset_index().drop(columns=["index",'educational-num'])
```

## category dataFrame ( str to num )

```python
for column in df.columns : 
    if type(df[column][0])==str:
        df[column] = LabelEncoder().fit(df[column].value_counts().index).transform(df[column].values)
```

```python
df.head()
```

## choose best var x 

```python
x = df.drop(columns=['income','fnlwgt','native-country']).values
y = df['income'].values
```

## Train/Test split

```python
train_x , test_x , train_y , test_y = train_test_split(x,y,test_size=0.2,random_state=4)
print(f'test : x size {test_x.shape} y size {test_y.shape}')
print(f'train : x size {train_x.shape} y size {train_y.shape}')
```

# Classification
## Decision Tree 

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=10)
tree.fit(train_x,train_y)
test_y_hat = tree.predict(test_x)
print(accuracy_score(test_y,test_y_hat))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test_y, test_y_hat)
plt.figure(figsize=(10, 8))
sns.heatmap(mat, annot=True)
plt.show()
```

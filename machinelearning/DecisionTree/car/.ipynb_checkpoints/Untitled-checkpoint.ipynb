## import Libs and Frameworks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

## read dataset

```python
df = pd.read_csv('car_evaluation.csv')
df.columns = ['buying price','maintenance cost','number of doors','number of persons','lug boot','safety','decision']
df.head()
```

## category dataFrame ( str to num )

```python
attributes_column = dict()
for column in df.columns:
    attributes_column[column] = dict(df[column].value_counts())
for attributs in attributes_column.items() :
    print(f"column '{attributs[0]}' have {len(attributs[1])} attributs :")
    for attribut in attributs[1].items() :
        print(f"\tattribut '{attribut[0]}' : count {attribut[1]}")
```

```python
df.shape
```

```python
for attributes in attributes_column.items() :
    df[attributes[0]] = LabelEncoder().fit(list(attributes[1].keys())).transform(df[attributes[0]])
```

## Plot data

```python
sns.set_theme()
ph = df.hist(figsize=(20,20))
```

```python
plt.figure(figsize=(15,15))
p = sns.heatmap(df.corr(), annot=True,cmap ='RdYlGn')
```

```python
sns.pairplot(df,hue='decision')
```

## choose best var x 

```python
x = df.drop(columns=['decision','lug boot','number of doors']).values
y = df[['decision']].values
df.head()
```

## Train/Test split

```python
train_x , test_x , train_y , test_y = train_test_split(x,y,test_size=0.3,random_state=4)
print(f'test : x size {test_x.shape} y size {test_y.shape}')
print(f'train : x size {train_x.shape} y size {train_y.shape}')
```

# Classification
## Decision Tree 

```python
drugTree = DecisionTreeClassifier(criterion='entropy',max_depth=4)
drugTree.fit(train_x,train_y)
test_y_hat = drugTree.predict(test_x)
print(accuracy_score(test_y,test_y_hat))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test_y, test_y_hat)
plt.figure(figsize=(10, 8))
sns.heatmap(mat, annot=True)
plt.show()
```
